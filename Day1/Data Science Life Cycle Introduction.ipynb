# Data Science Lifecycle: Herding Schrodinger's Cats
----
The Data Science Life Cycle is a structured format to bring to your projects

The lifecycle defines the steps needed to execute the project from start to finish. It can be baked into many other project management styles or life cycles. 

* THIS IS NOT THE THING TO USE FOR _EXPLORATORY DATA SCIENCE_ or _AD HOC_ types of data discovery.

* The Data Science Lifecycle is best suited for building data applications that produce data products and intelligent applications that include predictive analytics from Machine Learning Models or Artificial Intelligence models. 

```python
Image("../src/photos/build_maintian_deploy.png")
```

## There are 5 Major Steps to the lifecycle
---
1. Business Understanding
2. Data Acquisition and Understanding
3. Modeling
4. Deployment
5. Customer Acceptance

```python
Image("../src/photos/full_lifecycle.png")
```

## 1. Business Understanding
----
What we're trying to do here is build an understanding of the targets we want the model to focus on. We need to find out the types of questions we are trying to answer, are we able to answer the questions with the knowledge and techniques we have now.
Once we are able to locate data sources we need to choose the predictive models or bringing in other types of data from other sources.


### Two Main Steps:

1. Identify the variables you want to predict (*model targets*)

2. Identify the data sources you need to pull from 


Working with the customer and stakeholders involved to understand the business problems that can greatly enhance out ability to use preditive analytics. 


#### Define Your Objectives:
----

* Understand the customers business enviroment, their key variables, and how they are defining success. How can data science improve what's going on? How can it complicate things?

* Delphi meet Dephi: I can answer anything as long as the answer is a string or an int ;)~

    * How much or how many => _regressive techniques_
    * Which Category => _classification_
    * Which Group => _clustering_
    * Which one isn't like the others? => _Anomaly Detection_
    * Which  options would you like? => _recommendation engines_
    
* Define the project team, the role, and responsiblities. Develop a plan to *iterate* on as more information is discovered

* Define the success metrics. *SMART* metrics are _Specific, Measurable, Achievable, Relevant, and Time-Constrained



#### Identify Data Sources:
----

* Identify data sources that contain known examples of answers to the questions. 
* Data must be _*relevant*_ to the question. Are there measures and features related to the target??
* Data that is _*accurate*_ to the measure of the model target and the features of interest

It is not uncommon, for example, to find that existing systems need to collect and log additional kinds of data to address the problem and achieve the project goals. In this case, you may want to look for external data sources or update your systems to collect newer data.
Artifacts


#### Deliverables:
----

The following are the deliverables in this stage.

* *Charter Document* : Updated throughout the project as new discoveries are made and as _business requirements_ change. The key is to iterate upon this document with finer details.

* *Data Sources*: It describes the sources for the raw data. Fill in additional details like pseudocode to move the data to your analytic ernvironment.

* *Data Dictionaries*: Provides the descriptions and the schema (data types, validation rules). Entity-relation diagrams or descriptions are included too.

## 2 - Data Acquisition
----
* Ingest the data into the target analytic environment
* To determine if the data we have can be used to answer the question.

* Start developing the process to move the data from the source location to the target locations

* Real world data is often messy with incomplete or incorrect data. 

* By data summarization and visualization of the data, you can quickly identify the quality of your data and inform how to deal with the data quality

* Start to better understand the inherent patterns in the data that will help you develop a sound predictive model for your target.

* Specifically you look for evidence for how well connected is the data to the target and whether the data is large enough to move forward with next steps



### Data Acquisition: Pipelines
---
* Initial ingestion of data, 
* Setup a process to score new data or refresh the data
* Developed in parallel in the following stages of the data science project


#### Deliverables:
----

The following are the deliverables in this stage.



*    *Data Quality Report* :  data summaries, relationships between each attribute and target, variable ranking etc.

*   *Solution Architecture*: This can be a diagram and/or description of your data pipeline used to run scoring or predictions on new data once you have built a model. It will also contain the pipeline to retrain your model based on new data.

### _Checkpoint Decision_: 
---
Before we do the full feature engineering and model building; can we reevaluate the project to determine value in continuing this effort.

## 3- Modeling
----
* _Feature Engineering_: Develop new attributes or data features
* Construct and evaluate an informative model to predict the target.
* Determine if we have a model that is suitable for production use



### 3.1 Feature Engineering
----
* Feature engineering involves inclusion, aggregation and transformation of raw variables to create the features

* Need to understand how features are related to each other, and how the machine learning method will be using those features. 

* Requires a creative combination of domain expertise and the insights obtained from the data exploration step.

* balancing act of including informative variables without including too many unrelated variables.

### 3.2- Model Training
----

* Depending on type of question => there are multiple modeling algorithms 

* Input data is split randomly into a training data set and a test data set.
* Models are built using the training data set and evaluated

* Best solution by comparing the success metric between alternative methods.



```python
Image("../src/photos/model_training.png")
```

### [NOTE] Avoid leakage:
----

* Leakage is caused by including variables that can perfectly predict the target.

* These are usually variables that may have been used to detect the target initially.

* To avoid this often requires bouncing between building an analysis data set, creating a model and evaluating the accuracy.

#### Leakage is a major reason data scientists get nervous when they get really good predictive results.



#### Deliverables:
----

* _Feature Sets_: The features developed for modeling.
* _Modeling Report_: For each models that are tried

#### _Checkpoint Decision_: Evaluate whether the model performing enough to deploy 
----
* Does the model answer the question sufficiently given the test data?
* Should we go back and collect more data or do more feature engineering or try other algorithms?

### Deploy models and pipeline  production
----

* A set of models that perform can be operationalized for other applications to use.
* Depending on the business requirements, predictions are made either in real time or on a batch basis.
* To be operationalized, models have to be exposed with an open API interface 
* A good idea to build in telemetry and monitoring of the model and the data pipeline for system status reporting and troubleshooting.

* Final modeling report with deployment details
