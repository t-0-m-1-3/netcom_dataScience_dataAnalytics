### Default Parameters
----
* Default parameters are defined with `=` in the function signature
* If you do not pass an argument in, the default argument will be used.


````python
def whats_on_the_telly(penguin=None):
    if penguin is None:
        penguin = []
    penguin.append("Property of the zoo")
    return penguin
````

````python
# from help("def")
whats_on_the_telly()
````

### Simple Interest Calculator
----
We're going to write a function that will take in a principal amount, an interest rate ( as a float ) and the time the money will be borrowed for. 

It will output a simple interest rate

````python
def simple_interest(principal, rate, time):
    """
    This function returns the simple interest rate caluculated as ...
    """
    interest = principal * rate * time
    return interest
````

#### Kwargs and Args
----
Key word arguments and Arugments can give you a few special abilities
like taking in dictionaries of arguments and iterating overthem.


````python
def print_keyword_args(**kwargs):
    #kwargs is a dict of the arguments passed in
    for key,value in kwargs.items():
        print("%s = %s" % (key, value))

print_keyword_args(first_name='tom', last_name='goodheart')
````

#### From The Docs
----

````python
# from python official tutorial
def cheeseshop(kind, *arguments, **keywords):
    print("[+] Do you have any", kind, "?")
    print("[-] I'm sorry, we're all out of", kind)
    for arg in arguments:
        print(arg)
    print("-" * 40)
    for kw in keywords:
        print(kw, ":", keywords[kw])
````

````python
cheeseshop("Limburger", "It's very runny, sir.",
           "It's really very, VERY runny, sir.",
           shopkeeper="Michael Palin",
           client="John Cleese",
           sketch="Cheese Shop Sketch")
````

### Scope and Namespace:
----

A _namespace_ is a mapping from names to objects. 
    * currently implemented as Python dictionaries,

Examples:
* the set of built-in names (containing functions such as abs(), and built-in exception names) 
* the global names in a module; 
* the local names in a function invocation. 

The important thing to know about namespaces is that *there is absolutely no relation between names in different namespaces* 


### Scope is the textual region of a program where a namespace is accessible

````python
def scope_test():
    spam = 'after glboal'
    def do_local():
        spam = "local spam"

    def do_nonlocal():
        nonlocal spam
        spam = "nonlocal spam"

    def do_global():
        global spam
        spam = "global spam"

    spam = "test spam"
    do_local()
    print("After local assignment:", spam)
    do_nonlocal()
    print("After nonlocal assignment:", spam)
    print(spam)
    do_global()
    print("After global assignment:", spam)

scope_test()
print("In global scope:", spam)
````

### Methods
---
Methods are functions that are define under the classes scope

````python
class Circle:
    pi = 3.14
    
    #circle gets initiated with a radius
    def __init__(self, enter_radius=1):
        self.radius= enter_radius
        
    # Area method will calculate the area
    def area(self):
        return Circle.pi * self.radius ** 2
    
    # Method for resetting Radius
    def setRadius(self, raduis1):
        self.radius = raduis1
        
    # getting the radius
    def getRadius(self):
        return self.radius
````

````python
c = Circle(enter_radius=3)
````

````python
c.area()
````

````python
c.getRadius()
````

````python
c.setRadius(2)
c.area()
````

### Extending and Overriding:
----
* Base classes help form derived classes

* derived classes extend or overide the functionality of ancestors


````python
class Animal():
    def __init__(self):
        print("Animal created")
    
    def whoAmI(self):
        print("Animal")
    def eat(self):
        print("eating")
````

````python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")
    def whoAmI(self):
        print("Dog")
    def bark(self):
        print("woof")
````

````python
class Customer():
    """
    A customer of ABC Bank with a checking account. 
    Attributes:
        name: A string representing the customer's name.
        balance: A float tracking the current balance of the customer's account.
    """
    
    def __init__(self, name, balance=0.0):
        """
        Return a Customer object whose name is *name* and starting
        balance is *balance*.
        """
        self.name = name
        self.balance = balance
        
    def available_balance(self):
        """
        Return the available balance
        """
        return self.balance

    def withdraw(self,amount):
        """
        Return the balance remaining after withdrawing *amount*
        """
        if amount > self.balance:
            print('Amount greater than available balance.')
        else:
            self.balance -= amount
            return self.balance

    def deposit( self, amount):
        """
        Return the balance remaining after depositing *amount*
        """
        self.balance += amount
        return self.balance
````
