## Distribution plot
---
Flexibly plot a univariate distribution of observations.

```python
sns.kdeplot()
```

```python
# Show a default plot with a kernel density estimate and histogram with bin
# size determined automatically with a reference rule:

sns.distplot(tips.total_bill)
```

This function combines the matplotlib ``hist`` function (with automatic
calculation of a good default bin size) with the seaborn :func:`kdeplot`
and :func:`rugplot` functions. It can also fit ``scipy.stats``
distributions and plot the estimated PDF over the data.



```python
sns.distplot(tips.total_bill,  kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
tips.shape
```

```python
sns.distplot(tips["total_bill"], kde=False, bins=100)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips)
```

```python
sns.distplot(tips.tip,  kde=False)
```


Show a default plot with a kernel density estimate and histogram with bin size determined automatically with a reference rule:


```python
# kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }, optional
# Kind of plot to draw.

sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = "hex")
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = "reg")
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = "kde")
```

```python
# Plot pairwise relationships in a dataset
# this function will create a grid of Axes such that each
# variable in ``data`` will by shared in the y-axis across a row and
# in the x-axis across a single column. The diagonal Axes are treated
# differently, to show the univariate distribution of the data
# for the variable in that column
sns.pairplot(tips[['total_bill', 'tip']])
```

```python
sns.pairplot(tips[['total_bill', 'tip', 'sex']], hue = "sex")
```

```python
sns.boxplot(data = tips['total_bill'])
```

Draw a box plot to show distributions with respect to categories.

shows the distribution of quantitative data in a way to compare between variables or across levels of a categorical variable.

The box shows the _quartiles_ of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be "outliers" using a method that is a function of the _inter-quartile range_.



### Categorical Plot

```python
sns.barplot(x = "sex" , y = "total_bill" , data = tips)
```

### `barplot`
----

Show point estimates and confidence intervals as rectangular bars.

A bar plot represents an estimate of central tendency for a numeric
variable with the height of each rectangle and provides some indication of the uncertainty 
They are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.


```python
sns.barplot(x = "sex" , y = "total_bill" , data = tips, estimator= np.median)
```

```python
sns.countplot(x = "sex", data = tips)
```

A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. The basic API and options are identical to those for :func:`barplot`, so you can compare counts across nested variables.



```python
sns.boxplot(x ='day' , y ='total_bill' , data = tips)
```

```python
sns.boxplot(x ='day' , y ='total_bill' , data = tips, hue = "sex")
```

### Regression Plot
---
Plot data and regression model fits across a FacetGrid.

It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.


```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', col = "sex")
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', col = "sex", row = "time")
```

```python
titanic = sns.load_dataset("titanic")

g = sns.countplot(x="class", data=titanic)

# size of the figure
g.figure.set_size_inches(12,8)
 

# removing ticks upper and right
sns.despine() 

# title of the graph
g.axes.set_title("Distribution of Class", color = 'r', alpha = 0.5, fontsize = 34) 

# Setting the x-axis label
g.set_xlabel("Class", color = 'g', alpha = 0.5, fontsize = 25)

# Setting the x-axis label
g.set_ylabel("Count", color = 'g', alpha = 0.5, fontsize = 25)

# setting the ticks parameters
g.tick_params(labelsize = 14, labelcolor = "black")
```
