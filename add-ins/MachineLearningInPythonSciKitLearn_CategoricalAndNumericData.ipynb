* machine learning is an algorithm that can learn from data

#### Supervised vs. Unsupervised Learning Models
-----
* Supervised learning models are pre_labeled to help correct the model
* Unsupervised learning models are non-labled models; only have the data; structures are found; no _Y_ variables; no corpus; no training phase

* training phase 
* loss functions will help make the model better
* Prediction phase

* _X_ Data points are lists or vector of these features
* _Y_ labels are also called labels; these are what we are trying to predict. 

* Linear Regression is the most common used; we find the function that links all the features; y = mx + b

* Clustering is a very popular unsupervised learning technique


#### Continuos Data vs. Categorical Data
----
* Continuous -> infiitue set of values -> needs basic high school maths
* Categorical -> finite set of variables; just two variables _binary variables_



#### Mean
----
* the mean is the average of the points; it's a single number representing all the points and dividing by the total number

#### Variation
----
* Do the numbers jump around; and by how much? 
* Calculate the mean deviation of each data point
* Squares mean deviation just squares that result
* variance is the sum of the squared distance of the points from the mean, divided by the number of X; _Bessons's_ correction is `n-1`

#### Standard Deviations
----
* Square Root of variance


#### Range
----
* Range ignores the variance; it's just the max-min

#### Standardizing Data
-----
* subtract the mean and divide by the standard deviation of the numbers

* Standardizing data will cause data to have a mean of 0 and a variance of 1; 

* It prevents high variance data series from dominating the Models;
* PCA, Lasso, Ridge Regression

#### Check out average score for each exam

### Data Standardization:
Apply scaling on the test scores to express them in terms of <b>z-score</b> <br />
Z-score is the expression of a value in terms of the number of standard deviations from the mean <br />
The effect is to give a score which is relative to the the distribution of values for that column

#### Standardizing Data Using the `preprocessing.scale()`
-----


```python
from sklearn import preprocessing

exam_data[['math score']] = preprocessing.scale(exam_data[['math score']])
exam_data[['reading score']] = preprocessing.scale(exam_data[['reading score']])
exam_data[['writing score']] = preprocessing.scale(exam_data[['writing score']])
```

```python
exam_data
```

#### Explore averages after scaling

##### After Scaling: How does this Affect the Mean??
----
After scaling the mean is zero( or almost zero ) and the standard deviation is `1`

```python
math_average = exam_data['math score'].mean()
reading_average = exam_data['reading score'].mean()
writing_average = average = exam_data['writing score'].mean()

print('Math Avg: ', math_average)
print('Reading Avg: ', reading_average)
print('Writing Avg: ', writing_average)
```

#### Categorical Data

Continuous data can be ordered; not so with categorical data
There is special encoding preprocessing you need to do to data to accomplish this
Numerical encodings should never be ordered


##### `One Hot Hot Encoding` => each value represents on position in the array; 

| thing 1 | thing 2 |
|---|---|
| 1 | 0 |
| 0 | 1 | 

### Label Encoding:
Convert text values to numbers. These can be used in the following situations:
* There are only two values for a column in your data. The values will then become 0/1 - effectively a binary representation
* The values have relationship with each other where comparisons are meaningful (e.g. low<medium<high)
* Invoke it with `preprocessing.LabelEncoder()` and use the `fit_transforma()` 

```python
help(preprocessing.LabelEncoder.fit_transform)
```

```python
le = preprocessing.LabelEncoder()
exam_data['gender'] = le.fit_transform(exam_data['gender'].astype(str))
```

```python
exam_data.head()
```

#### See how the data was transformed by categorical labeling!
---- 
I can still access my category labels using the `classes_`

```python
le.classes_
```

### One-Hot Encoding:
* Use when there is no meaningful comparison between values in the column
* Creates a new column for each unique value for the specified feature in the data set

```python
pd.get_dummies(exam_data['race/ethnicity'])
```

#### Include the dummy columns in our data set

```python
exam_data = pd.get_dummies(exam_data, columns=['race/ethnicity'])
```

```python
exam_data
```

#### Apply one-hot-encoding for remaining non-numeric features

```python
exam_data = pd.get_dummies(exam_data, columns=['parental level of education', 
                                               'lunch', 
                                               'test preparation course'])
```
